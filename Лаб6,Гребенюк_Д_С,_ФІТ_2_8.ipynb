Завдання: Побудувати інтерполяційний багаточлен Лагранжа (х) для функції f(х), що задана таблицею, та з точністю до 0.001 обчислити наближені значення функції у заданих точках. Побудувати графік інтерполяційної функції у = L (х) за наявним набором точок.

```python
import numpy as np
import matplotlib.pyplot as plt

x_data = np.array([-2, -1, 2])
y_data = np.array([-22, -10, -10])

x_interp = np.array([-1.5, -0.5, 1.5, 2.5])

def lagrange_interpolation(x, x_data, y_data):
    """Обчислює значення полінома Лагранжа в точці x"""
    n = len(x_data)
    y = 0
    for i in range(n):
        term = y_data[i]
        for j in range(n):
            if j != i:
                term *= (x - x_data[j]) / (x_data[i] - x_data[j])
        y += term
    return y

y_interp = [lagrange_interpolation(xi, x_data, y_data) for xi in x_interp]

plt.plot(x_data, y_data, 'o', label='Дані')
plt.plot(x_interp, y_interp, '-', label='Інтерполяція')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print("Значення функції в точках для інтерполяції:")
for x, y in zip(x_interp, y_interp):
    print(f"x = {x:.2f}, y = {y:.3f}")
```
