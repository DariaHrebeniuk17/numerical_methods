Розв'язання систем нелінійних рівнянь.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def f1(x, y):
    return np.sin(x + 0.5) - y - 1

def f2(x, y):
    return np.cos(y - 2) + x

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
F1 = f1(X, Y)
F2 = f2(X, Y)

plt.contour(X, Y, F1, levels=[0], colors='blue')
plt.contour(X, Y, F2, levels=[0], colors='red')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

x0 = [0.5, -1]

def iteration_method(x0):
    tol = 0.001
    max_iter = 100
    for i in range(max_iter):
        x1 = -np.cos(x0[1] - 2)
        y1 = np.sin(x1 + 0.5) - 1
        if np.linalg.norm([x1 - x0[0], y1 - x0[1]]) < tol:
            break
        x0 = [x1, y1]
    return x0

solution = iteration_method(x0)
print("Розв'язок:", solution)

sol = fsolve(lambda x: [f1(x[0], x[1]), f2(x[0], x[1])], x0)
print("Розв'язок за допомогою fsolve:", sol)
```
